```python
import pandas as pd
```

Import election data from the August 2023 Issue one (60% vote to ratify constitution)

```python
i1_aug_2023_df = pd.read_csv('./issue1_aug_2023.csv')
i1_aug_2023_df
```

Add a column to show the +/- of Yes versus No votes

```python
i1_aug_2023_df['No +/-'] = (i1_aug_2023_df['No'] / (i1_aug_2023_df['Yes'] + i1_aug_2023_df['No'])) - .5
i1_aug_2023_df
```

```python
pres_nov_2020 = pd.read_csv('./president_nov_2020.csv')
pres_nov_2020
```

```python
pres_nov_2020['Dem +/-'] = pres_nov_2020['Joseph R. Biden and Kamala D. Harris (D)'] / ((pres_nov_2020['Donald J. Trump and Michael R. Pence (R)'] + pres_nov_2020['Joseph R. Biden and Kamala D. Harris (D)']))-.5
pres_nov_2020
```

```python
%%capture
%pip install seaborn
import seaborn as sns
```

```python
pres2020 = pres_nov_2020[['County Name', 'Dem +/-']]
pres2020['2020 Turnout'] = pres_nov_2020['Official Voter Turnout']
issue1_2023 = i1_aug_2023_df[['County Name', 'No +/-']]
issue1_2023['2023 Turnout'] = i1_aug_2023_df['Official Voter Turnout']

df = pd.merge(pres2020, issue1_2023, on='County Name')
sns.lmplot(
    data=df, x="Dem +/-", y="No +/-"
)
```

The more left leaning the county, the more they voted no on issue 1. But a number of counties flipped.

Lets see if we can learn anything from how counties trended

```python
df['Change +/-'] = df['No +/-'] - df['Dem +/-']
df
```

```python
sns.lmplot(
    data=df, y="Change +/-", x="Dem +/-"
)
```

This is interesting because there is no correlation between how left leaning the county was and how it trended during the august issue 1 vote

```python
sns.lmplot(
    data=df, y="Change +/-", x="2023 Turnout"
)
sns.lmplot(
    data=df, y="Change +/-", x="No +/-"
)
```

Theres little to no correlation that turnout and results were indicators of the shift in results.
